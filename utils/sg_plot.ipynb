```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

# Sigmoid
sigmoid = 1 / (1 + np.exp(-x))
plt.figure(figsize=(6, 4))
plt.plot(x, sigmoid, label='Sigmoid', color='blue')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.savefig('sigmoid.png', dpi=300, bbox_inches='tight')
plt.close()

# ReLU
relu = np.maximum(0, x)
plt.figure(figsize=(6, 4))
plt.plot(x, relu, label='ReLU', color='green')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.grid(True)
plt.legend()
plt.savefig('relu.png', dpi=300, bbox_inches='tight')
plt.close()

# Tanh
tanh = np.tanh(x)
plt.figure(figsize=(6, 4))
plt.plot(x, tanh, label='Tanh', color='red')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()
plt.savefig('tanh.png', dpi=300, bbox_inches='tight')
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
alpha_leaky = 0.01
alpha_prelu = 0.2

leaky_relu = np.where(x > 0, x, alpha_leaky * x)
prelu = np.where(x > 0, x, alpha_prelu * x)

plt.figure(figsize=(6, 4))
plt.plot(x, leaky_relu, label='Leaky ReLU (=0.01)', color='blue')
plt.plot(x, prelu, label='PReLU (=0.2)', color='orange', linestyle='--')
plt.plot(x, np.maximum(0, x), label='ReLU', color='green', linestyle=':')
plt.title('Leaky ReLU and PReLU Activation Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.savefig('leaky_prelu.png', dpi=300, bbox_inches='tight')
plt.show()
```
